```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df["PrecioActual"], bins=50, color='skyblue', edgecolor='black')
plt.title("Distribución del Precio Actual")
plt.xlabel("Precio (€)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

```python
df= df[df["PrecioActual"] < 6000000]
```

```python
plt.figure(figsize=(10,6))
plt.hist(df["PrecioActual"], bins=50, color='skyblue', edgecolor='black')
plt.title("Distribución del Precio Actual")
plt.xlabel("Precio (€)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

```python
y = np.log1p(df["PrecioActual"])
```

```python
plt.figure(figsize=(10,6))
plt.hist(y, bins=50, color='skyblue', edgecolor='black')
plt.title("Distribución del Precio np log1p")
plt.xlabel("Precio log (€)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```
